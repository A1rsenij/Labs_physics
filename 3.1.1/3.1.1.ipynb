```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('3.6.1.xlsx', sheet_name='_')
names = list(data)

tau = data[names[1]]
dv = data[names[2]] * 1000

k, cov = np.polyfit(tau, dv, 1, cov=True)
dv_p = np.polyval(k, tau)

fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

ax.scatter(tau, dv, label =r"$\Delta{\nu}(1/\tau)$", marker='.')
ax.plot(tau, dv_p, label=r"$\Delta{\nu} = $" + r"{:.2f}$\pm${:.2f}".format(k[0], np.sqrt(cov[0][0])) + r"$\,\cdot\, \frac{1}{\tau}$") # + r" + {:.2f}".format(k[1]))

ax.legend(loc='upper left', fontsize=10)

ax.grid(which="major", linewidth=0.5)
ax.grid(which="minor", linestyle='--', linewidth=0.25)
plt.minorticks_on()

ax.axis([0, 55000, 0, 55000])

ax.set_title(r"  $\Delta{\nu}(1/\tau)$", loc='center', fontsize=15)
ax.set_xlabel(r"$1/\tau,\, 1/c$", loc='center', fontsize=10)
ax.set_ylabel(r"$\Delta{\nu},\, $", loc='center', fontsize=10)

plt.show()

fig.savefig("3.6.1_1.png", dpi=500)
```

```python
data = pd.read_excel('3.6.1.xlsx', sheet_name='')
names = list(data)

T = 1000 / data[names[0]][1:]
dv = data[names[1]][1:]

k, cov = np.polyfit(T, dv, 1, cov=True)
dv_p = np.polyval(k, T)

fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

ax.scatter(T, dv, label =r"$\delta{\nu}(1/T)$", marker='.')
ax.plot(T, dv_p, label=r"$\delta{\nu} = $" + r"{:.2f}$\pm${:.2f}".format(k[0], np.sqrt(cov[0][0])) + r"$\,\cdot\, \frac{1}{T}$") # + r" + {:.2f}".format(k[1]))

ax.legend(loc='upper left', fontsize=10)

ax.grid(which="major", linewidth=0.5)
ax.grid(which="minor", linestyle='--', linewidth=0.25)
plt.minorticks_on()

ax.axis([0, 1000, 0, 1000])

ax.set_title(r"  $\delta{\nu}(1/T)$", loc='center', fontsize=15)
ax.set_xlabel(r"$1/T,\, 1/c$", loc='center', fontsize=10)
ax.set_ylabel(r"$\delta{\nu},\, $", loc='center', fontsize=10)

plt.show()

fig.savefig("3.6.1_2.png", dpi=500)
```

```python
data = pd.read_excel('3.6.1.xlsx', sheet_name='')
names = list(data)

m = data[names[0]]
a = data[names[5]]

k, cov = np.polyfit(m, a, 1, cov=True)
a_p = np.polyval(k, m)

fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

ax.scatter(m, a, label =r"$\frac{a_{}}{a_{}}\left(m\right)$", marker='.')
ax.plot(m, a_p, label=r"$\frac{a_{}}{a_{}} = $" + r"{:.3f}$\pm${:.3f}".format(k[0], np.sqrt(cov[0][0])) + r"$\,\cdot\, m$") # + r" + {:.2f}".format(k[1]))

ax.legend(loc='upper left', fontsize=10)

ax.grid(which="major", linewidth=0.5)
ax.grid(which="minor", linestyle='--', linewidth=0.25)
plt.minorticks_on()

ax.axis([0, 1.1, 0, 0.8])

ax.set_title(r"  $\frac{a_{}}{a_{}}\left(m\right)$", loc='center', fontsize=15)
ax.set_xlabel(r"$m$", loc='center', fontsize=10)
ax.set_ylabel(r"$a_{}/a_{}$", loc='center', fontsize=10)

plt.show()

fig.savefig("3.6.1_3.png", dpi=500)
```

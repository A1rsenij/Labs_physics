```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('3.4.2.xlsx', sheet_name='Лист1')
names = list(data)

y = np.array(data[names[6]])
x = np.array(data[names[8]])

xp = x[4:]
yp = y[4:]

k, cov = np.polyfit(xp, yp, 1, cov=True)
xp = np.insert(xp, 0, -k[1]/k[0]) 
yp_p = np.polyval(k, xp)
print("Точка Кюри:", xp[0], "+-", ((np.sqrt(cov[0][0])/k[0])**2 + (np.sqrt(cov[1][1])/k[1])**2) * -k[1]/k[0] + 0.026)


fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

ax.scatter(x, y, label =r"$\frac{1}{\tau^2-\tau_0^2} = f(T)$", marker='.')
ax.plot(xp, yp_p, linestyle='--', color='red')

ax.legend(loc='upper left', fontsize=10)

ax.grid(which="major", linewidth=0.5)
ax.grid(which="minor", linestyle='--', linewidth=0.25)
plt.minorticks_on()

ax.axis([0, 45, 0, 0.5])

ax.set_title(r"График зависимости $\frac{1}{\tau^2-\tau_0^2} = f(T)$", loc='center', fontsize=15)
ax.set_xlabel(r"$T,\, ^{\circ} C$", loc='center', fontsize=10)
ax.set_ylabel(r"$f(T),\, 1/мкс^2$", loc='center', fontsize=10)

plt.show()

fig.savefig("3.4.2_1.png", dpi=500)
```

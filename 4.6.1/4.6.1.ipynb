```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data = pd.read_excel('4.6.1.xlsx', sheet_name='Малюс')
names = list(data)

alpha = np.cos(np.deg2rad(2 * data[names[1]])) ** 2
I = data[names[2]]

fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

ax.scatter(alpha, I, label=r"$I(\cos^2{(\alpha)})$")

ax.legend(loc='upper left', fontsize=10)

ax.grid(which="major", linewidth=0.5)
ax.grid(which="minor", linestyle='--', linewidth=0.25)
plt.minorticks_on()

ax.axis([0.85, 1.01, -1, 85])

ax.set_title(r"Зависимость уровня сигнала $I$ от угла поврота приёмной антенны $\alpha$ относительно луча", loc='center', fontsize=15)
ax.set_xlabel(r"$\cos^2{(\alpha)}$", loc='center', fontsize=10)
ax.set_ylabel(r"$I,\; мкВ$", loc='center', fontsize=10)

plt.show()

fig.savefig("4.6.1_1.png", dpi=500)
```

```python
def func1(x, i1, i2, k):
    return (i1 + 2 * np.sqrt(i2) * np.cos(k * x))
```

```python
data = pd.read_excel('4.6.1.xlsx', sheet_name='Интерф')
names = list(data)

x = data[names[0]]
I = data[names[1]]

k, cov = curve_fit(func1, x, I)
I_p = func1(x, k[0], k[1], k[2])

fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

ax.scatter(x, I, label=r"$I(x)$")
ax.plot(x, I_p, label=rf"$I(x) = {k[0]:.3f} + 2 \cdot sqrt ({(k[1]):.3f}) \cdot cos ({-k[2]:.3f} \cdot x)$")

ax.legend(loc='upper left', fontsize=10)

ax.grid(which="major", linewidth=0.5)
ax.grid(which="minor", linestyle='--', linewidth=0.25)
plt.minorticks_on()

ax.axis([-0.2, 3, 0, 100])

ax.set_title(r"Зависимость интенсивности $I$ от координаты $x$ подвижного зеркала", loc='center', fontsize=15)
ax.set_xlabel(r"$x,\; мм$", loc='center', fontsize=10)
ax.set_ylabel(r"$I,\; мкВ$", loc='center', fontsize=10)

plt.show()

fig.savefig("4.6.1_2.png", dpi=500)
```

```python
def func2(x, k):
    return (k * x)
```

```python
data = pd.read_excel('4.6.1.xlsx', sheet_name='Майкельсон')
names = list(data)

m = data[names[0]]
x_m = data[names[2]]

k, cov = curve_fit(func2, m, x_m)
x_m_p = func2(m, k[0])

fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

ax.scatter(m, x_m, label=r"$x_m = \mathit{f}(m)$")
ax.plot(m, x_m_p, label=rf"$x_m = {k[0]:.3f} \cdot m$")

ax.legend(loc='upper left', fontsize=10)

ax.grid(which="major", linewidth=0.5)
ax.grid(which="minor", linestyle='--', linewidth=0.25)
plt.minorticks_on()

#ax.axis([-0.2, 3, 0, 100])

ax.set_title(r"Зависимость координаты интерференционного максимума $x_m$ от его номера $m$", loc='center', fontsize=15)
ax.set_xlabel(r"$m$", loc='center', fontsize=10)
ax.set_ylabel(r"$x_m,\; мм$", loc='center', fontsize=10)

plt.show()

fig.savefig("4.6.1_3.png", dpi=500)
```

```python
data = pd.read_excel('4.6.1.xlsx', sheet_name='Интенс')
names = list(data)

x = data[names[1]]
I = data[names[2]]

i1, i2 = 10, 14

I_d = i1 + i2 + 2 * np.sqrt(i1 * i2) * np.cos(2 * np.pi / 8.15 * np.linspace(0, 8.5, 10))

fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

ax.scatter(2 * x, I, label=r"$I(\Delta)$")
ax.plot(np.linspace(0, 8.5, 10), I_d, label=r"$I = I_1 + I_2 + 2 \cdot \sqrt{I_1 I_2} \cdot \cos{(k \cdot 2\delta{x})}$")

ax.legend(loc='upper left', fontsize=10)

ax.grid(which="major", linewidth=0.5)
ax.grid(which="minor", linestyle='--', linewidth=0.25)
plt.minorticks_on()

#ax.axis([-0.2, 3, 0, 100])

ax.set_title(r"Зависимость интенсивности $I$ от разности хода $\Delta$", loc='center', fontsize=15)
ax.set_xlabel(r"$\Delta,\; мм$", loc='center', fontsize=10)
ax.set_ylabel(r"$I,\; мкВ$", loc='center', fontsize=10)

plt.show()

fig.savefig("4.6.1_4.png", dpi=500)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
theta = np.arange(0, 120+1, 10)
N = np.array([929, 928, 833, 783, 692, 620, 546, 478, 437, 220, 196, 186, 174])

x = 1 - np.cos(np.deg2rad(theta))
y = 1 / N
err = 0.005 / 2 * np.ones(N.shape)

k, cov = np.polyfit(x, y, 1, cov=True)
yp = np.polyval(k, x)

plt.style.use('seaborn-v0_8')

plt.scatter(x, y, s=25, zorder=5, marker='o', label=r"$\frac{1}{N(\theta)} = A(1 - \cos{\theta}) + \frac{1}{N(0)}$")
plt.errorbar(x, y, xerr=0, yerr=err, fmt='none', zorder=4)
plt.plot(x, yp, linestyle='--', color='grey', zorder=3, label=r"$\frac{1}{N(\theta)} = $" + f"{k[0]:.5f}$\pm${np.sqrt(cov[0][0]):.5f}" + r"$\cdot (1 - \cos{\theta}) + $" + f"{k[1]:.5f}$\pm${np.sqrt(cov[1][1]):.5f}")

plt.legend(loc="upper left")
plt.ylabel(r'$\frac{1}{N(\theta)}$')
plt.xlabel(r'$1 - \cos{\theta}$')
plt.title(r"Зависимость номера канала фотопика от угла отклонения")

plt.show()
```

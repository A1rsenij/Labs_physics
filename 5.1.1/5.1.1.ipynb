```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
l = np.array([5401, 5945, 6074, 5267, 6533, 6717]) * 1e-10
V0 = np.array([0.745, 0.681, 0.601, 0.541, 0.441, 0.413])

w = 2 * np.pi * 3e8 / l

err = 0.1 * V0

k, cov = np.polyfit(w, V0, 1, cov=True)
Vp = np.polyval(k, w)

plt.style.use('seaborn-v0_8')

plt.scatter(w, V0, s=25, zorder=5, marker='o', label=r"$V_0(\omega)$")
plt.errorbar(w, V0, xerr=0, yerr=err, fmt='none', zorder=4)
plt.plot(w, Vp, linestyle='--', color='grey', zorder=3, label=r"$V_0 = $" + f"{k[0]:.3e}$\pm${np.sqrt(cov[0][0]):.3e}" + r"$\cdot \omega + $" + f"{k[1]:.3f}$\pm${np.sqrt(cov[1][1]):.3f}")

plt.legend(loc="upper left")
plt.ylabel(r'$V_0,,\; В$')
plt.xlabel(r'$\omega,\; c^{-1}$')
plt.title(r"Зависимость запирающего потенциала от частоты света")

plt.show()
```

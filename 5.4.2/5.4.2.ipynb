## Теоретические сведения

### Бета-распад

**Бета-распадом называется самопроизвольное превращение ядер, при котором их массовое число не изменяется, а заряд увеличивается или уменьшается на единицу.** В нашей работе происходит электронный распад, в котором кроме электрона испускается антинейтрино:

$$
{ ^{A}_{Z}X -> ^{A}_{Z+1}X + e^- + \widetilde{\nu} }
$$

Почти вся энергия, высвобождающаяся в данном распаде, делится между электроном и антинейтрино. Следовательно, электроны могут иметь любое значение энергии — от нулевого до некоторого максимального знечения энергии, выделяющейся в $\beta$-распаде. Пусть $p$ и $k$ — импульсы электрона и антинейтрино соответственно. Закон созранения энергии в таком случае выглядит так:

$$
E_{e} - E - ck = 0,
$$

где $E_{e}$ — максимальная энергия электрона, кинетическая энергия $E$ связана с импульсом $p$ соотношением

$$
E = c \sqrt{p^2 + m^2 c^2} - m c^2,
$$

а через $c k$ обозначена энергия антинейтрино с импульсом $k$.

### Распределение импульсов электрона

Через $\mathrm{d} \omega$ обозначим вероятность того, что при распаде электрон вылетит с импульсом $\mathrm{d}^3 p$, а антинейтрино — с импульсом $\mathrm{d}^3 k$. Величина $\mathrm{d} \omega$ будет пропорциональна произведению этих дифференциалов. Чтобы учесть закон сохранения энергии, введём для $\mathrm{d} \omega$ $\delta$-функцию:

$$
\delta (E_{e} - E - ck).
$$

По определению, она не равна нулю только при соблюдении условия ЗСЭ. В итоге получаем

$$
\mathrm{d} \omega = D \delta (E_{e} - E - ck) \mathrm{d}^3 p \mathrm{d}^3 k = D \delta (E_{e} - E - ck) p^2 \mathrm{d} p k^2 \mathrm{d} k \mathrm{d} \Omega_{e} \mathrm{d} \Omega_{\widetilde{\nu}}.
$$

Здесь $D$ — некоторый коэффициент пропорциональности, который мы можем счиать константой, а $\mathrm{d} \Omega_{e}$ и $\mathrm{d} \Omega_{\widetilde{\nu}}$ — элементы телесных углов вылета электрона и антинейтрино соответственно.

Для большого числа $N_0$ распадов число $\mathrm{d} N$ распадов, при которых импульсы электрона и антинейтрино находятся соответственно в интервалах от $p$ до $p + \mathrm{d} p$ и от $k$ до $k + \mathrm{d} k$, определяется соотношением

$$
\mathrm{d} N = N_0 \mathrm{d} \omega.
$$

$\mathrm{d} \omega$ можно проинтегрировать по всем телесным углам и по абсолютному значению импульса антинейтрино. Интегрирование по каждому телесному углу даёт множитель $4 \pi$, а интегрирование по импульсу антинейтрино использует свойство $\delta$-функции Дирака. По определению:

$$
\int\limits^{+ \infty}_{- \infty} \delta (x) \mathrm{d} x = 1 \Rightarrow \int\limits^{+ \infty}_{- \infty} f (x) \delta (x) \mathrm{d} x = f(0).
$$

В итоге при интегрировании по $k$ $\delta$-функция исчезнет, следовательно $c k$ всюду можно будет заменить на $(E_{e} - E)$. Умножим результат интегрирования на полное число распадов $N_0$ и получим число электронов, вылетающих из ядра с импульсом в интервале от $p$ до $p + \mathrm{d} p$:

$$
\mathrm{d} N = \frac{16 \pi^2 N_0}{c^2} D p^2 (E_{e} - E)^2 \mathrm{d} p.
$$

### Распределение энергий электрона

Теперь мы хотим получить распределение по энергиям. Для этого перейдём от $\mathrm{d} p$ к $\mathrm{d} E$:

$$
\mathrm{d} E = \frac{c^2 p}{E + m c^2} \mathrm{d} p.
$$

Форму $\beta$-спектра выражает величина $N (E) = \mathrm{d} N / \mathrm{d} E$:

$$
\frac{\mathrm{d} N}{\mathrm{d} E} = N_0 B c p (E + m c^2) (E_{e} - E)^2 = N_0 B \sqrt{E (E + 2 m c^2)} (E + m c^2) (E_{e} - E)^2,
$$

где $B = (16 \pi^2 / c^4) D$. В нерелятивистском приближении выражение выше упрощается и приводится к виду

$$
\frac{\mathrm{d} N}{\mathrm{d} E} \approx \sqrt{E} (E_{e} - E)^2.
$$

### Конверсия

Также в результате $\beta$-распада дочерние ядра нередко оказываются возбуждёнными. Возбуждённые ядра отдают энергию либо излучая $\gamma$-квант (энергия которого равна разности энергии начального и конечного уровней), либо передавая избыток энегрии одному из электронов внутренних оболочек атома. Излучаемые в таком процессе электроны имеют строго определённую энергию и называются конверсионными. На спектре появляется монохроматическая линия, соответствующая энергии электронов конверсии. Ширина линии определяется исключительно разрешающей способностью спектрометра.

## Методика измерений

Для определения энергии $\beta$-частиц в работе используется магнитный спектрометр, схема которого показана на рисунке.

<img src="Spectrometer.png" alt="spectrometer" width="400"/>

Электроны испускаются радиоактивным источником и попадают в магнитное поле катушки, ось которой параллельна $OZ$. Траектории электронов сходятся в одной точке — фокусе, где и установлен сцинтилляционный счетчик, сигналы которого усиливаются фотоумножителем и регистрируются пересчетным прибором. Фокусное расстояние $f$ магнитной линзы связано с током в катушке $I$ и импульсом $p_{e}$ регистрируемых частиц следующим образом:

$$
\frac{1}{f} \propto \frac{I^2}{p_{e}^2}.
$$

При неизменной геометрии установки, увеличивая и уменьшая силу тока, можно фокусировать электроны разных импульсов, причем 
	
$$
p_{e} = k I,
$$

где $k$ — коэффициент пропорциональности, являющийся параметром установки.

Схема установки для изучения спектра:

<img src="Scheme.png" alt="scheme" width="500"/>

В $\beta$-спектрометре установлены диафрагмы для ограничения углов вылета частиц из источника и свинцовый фильтр для защиты от прямого попадания $\gamma$-лучей. 

Число частиц $N$, регистрируемых на установке, равно: $N \approx W \cdot \Delta p_{e}$, где $\Delta p_{e}$ — разрешающая способность спектрометра. Дифференцируя выражение для форуса магнитной линзы, получим: $\Delta p_{e} = \frac{1}{2} \frac{\Delta f}{f} p_{e}$, то есть $\Delta p_{e} \propto p_{e}$. Таким образом, для количества частиц справедлива формула: 

$$
 N = C W (p_{e}) p_{e}. 
$$

Здесь $C$ - некоторая константа.

2. На графике построим зависимость числа частиц за вычетом фона $N - N_\text{Ф}$ от тока магнитной линзы $I$. Данная зависимость должна оппроксимироваться функциями вида

$$
y(x) = d + a \exp\left( \dfrac{(x - b)^2}{2c^2} \right)
$$
для конверсионного пика, и
$$
y(x) = c + a \cdot x^2 \cdot (b - \sqrt{x^2 + d})^2
$$
для $\beta$-спектра.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def func1(x, a, b, c, d):
    return d + a*np.exp(-(x - b)**2 / (2*c**2))

def func2(x, a, b, c, d):
    return c + a * x**2 * (b - np.sqrt(x**2 + d))**2
```

```python
I = np.array([0.0, 0.22, 0.41, 0.61, 0.80, 1.00, 1.20, 1.40, 1.60, 1.80, 2.02, 2.20, 2.40, 2.60, 2.80, 3.00, 3.20, 3.40, 3.60, 3.80, 3.85, 3.90, 4.00, 4.05, 4.10, 4.15, 4.20, 4.30, 4.35, 4.40, 4.45, 4.60, 4.80, 5.00])
NN_f = np.array([-0.143, 0.017, -0.123, 0.077, 0.177, 0.327, 1.116, 1.726, 2.416, 2.606, 3.476, 3.716, 3.815, 3.182, 2.815, 2.482, 1.732, 1.166, 0.383, 0.583, 1.482, 2.615, 3.715, 4.348, 5.714, 5.264, 4.598, 4.314, 2.165, 1.016, 0.933, -0.333, -0.450, -0.317])

N_ERR = np.abs(NN_f * 0.1)

k, cov = curve_fit(func1, I[np.where((I > 3.5) & (I < 5))], NN_f[np.where((I > 3.5) & (I < 5))])
print(k)
print(np.sqrt(cov))
N_p1= func1(I[np.where((I > 3.5) & (I < 5))], *k)

k, cov = curve_fit(func2, I[np.where((I > 0) & (I < 4))], NN_f[np.where((I > 0) & (I < 4))])
print(k)
N_p2 = func2(I[np.where((I > 0) & (I < 4))], *k)

plt.style.use('seaborn-v0_8')

plt.scatter(I, NN_f, s=25, zorder=5, marker='.')
plt.errorbar(I, NN_f, xerr=0, yerr=N_ERR, fmt='none', zorder=4)
plt.plot(I[np.where((I > 3.5) & (I < 5))], N_p1, linestyle='--', color='grey', zorder=3)
plt.plot(I[np.where((I > 0) & (I < 4))], N_p2, linestyle='--', color='grey', zorder=3)

plt.xlabel('$I$, A')
plt.ylabel('$N - N_ф$, 1 / sec')
plt.title(r"Измерение $\beta$-спектра")

plt.show()
```

С помощью параметров аппроксимации можно определить коэффициент $k$:
$$
p_c = 1013~кэВ = kI_0,
$$
где $I_0 = 4,13\pm0,01 A$ - сила тока, при которой наблюдается конверсионный пик. Получаем
$$ k = (245,39\pm0,59)\frac{кэВ}{с \cdot А} $$

Зная конверсионный пик и соответствующие ему импульс $ p_c = 1013 $ кэВ/$ с $ и энергию $ T = 634 $ кэВ, мы можем откалибровать шкалу токов в шкалу импульсов и энергий.

Запишем соотношение для числа электронов с импульсами в интервале от $p$ до $p + dp$ в виде:

$$
N(p) \approx p^3 (E_e - E)^2 \Rightarrow \dfrac{\sqrt{N}}{p^{3/2}} \varpropto T_{max} - T
$$

Отложив по оси ординат величину $ mkFermi = \dfrac{\sqrt{N}}{p^{3/2}} \cdot 1e6 = f $, а по оси абсцисс - кинетическую энергию, мы можем построить график, называемый графиком Ферми-Кюри, и определить по нему $ T_{max} $ - в этих осях спектр $\beta$-распада описывается прямой, который мы можем аппроксимировать $ y = ax +b $.


```python
T = np.array([0, 2.9, 9.9, 21.6, 36.7, 56.3, 79.3, 105.5, 134.3, 165.5, 202.2, 238.8, 270.4, 308.3, 347.3, 387.4, 428.3, 470.0, 512.4, 555.3, 566.1, 577.0, 598.8, 609.7, 620.7, 631.7, 642.8, 664.9, 676.0, 687.1, 698.3, 731.8, 776.9, 822.3])
f = np.array([0, 324.2068, 0, 150.3952, 151.9758, 147.8620, 207.9496, 205.1995, 198.6966, 172.9400, 168.0021, 152.8280, 135.9033, 110.0751, 92.6473, 78.4407, 59.4857, 44.5599, 23.4410, 26.6655, 41.6987, 54.3224, 62.3298, 66.1862, 74.4908, 70.2106, 64.4482, 60.2673, 41.9625, 28.2550, 26.6171, 0, 0, 0])

delta1, delta2 = 0.05, -20
k, cov= np.polyfit(T[np.where((T > 100) & (T < 500))], f[np.where((T > 100) & (T < 500))], 1, cov=True)
print(k, np.sqrt(cov), sep="\n")
print("Tmax =", k[1]/-k[0], "+-", np.sqrt(((cov[0][0] + delta1**2)/k[0]**2) + ((cov[1][1] + delta2**2)/k[1]**2)) *  k[1]/-k[0])
f_p = np.polyval(k, np.linspace(0, 600, 50))

f_p1 = np.polyval([k[0] - delta1, k[1] - delta2], np.linspace(0, 600, 50))
f_p2 = np.polyval([k[0] + delta1, k[1] + delta2], np.linspace(0, 600, 50))

f_err = np.abs(f * 0.05)

plt.style.use('seaborn-v0_8')

plt.scatter(T, f, s=25, zorder=5, marker='.')
plt.errorbar(T, f, xerr=0, yerr=f_err, fmt='none', zorder=4)
plt.plot(np.linspace(0, 600, 50), f_p, linestyle='--', color='grey', zorder=3)
plt.plot(np.linspace(0, 600, 50), f_p1, linestyle='-', color='red', zorder=3)
plt.plot(np.linspace(0, 600, 50), f_p2, linestyle='-', color='red', zorder=3)

plt.xlabel('$T$, кэВ')
plt.ylabel(r'$mkFermi$, $\frac{c}{м^{3/2}}$')
plt.title(r"График Ферми")

plt.show()
```

В результате аппроксимации получаем, что при $ mkFermi = 0 $ можно найти $ T_{max} = \dfrac{b}{-a} \approx (560 \pm 77) \; кэВ $.
